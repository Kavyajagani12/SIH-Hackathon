```python
best_aic = float("inf")
best_order = None
results = []

for p in range(0, 4):
    for d in range(0, 3):
        for q in range(0, 4):
            try:
                model = ARIMA(train, order=(p,d,q), enforce_stationarity=False, enforce_invertibility=False)
                model_fit = model.fit()
                aic = model_fit.aic
                results.append((p,d,q,aic))
                if aic < best_aic:
                    best_aic = aic
                    best_order = (p,d,q)
            except:
                continue

print(f"🔎 Best ARIMA order: {best_order} with AIC={best_aic}")
```

```python
from supabase import create_client
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Initialize Supabase client
SUPABASE_URL="https://udkqseribwkkloktyozc.supabase.co"
SUPABASE_KEY="eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpc3MiOiJzdXBhYmFzZSIsInJlZiI6InVka3FzZXJpYndra2xva3R5b3pjIiwicm9sZSI6ImFub24iLCJpYXQiOjE3NTgxMDQyNDQsImV4cCI6MjA3MzY4MDI0NH0.aT1yeG9sPK1RLsoMLCfDLfpUS2bagH8FoeByTKC5UqE"
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

def fetch_water_levels():
    """Fetch water level data from Supabase."""
    response = supabase.table("water_levels").select("station_id, timestamp, water_level").order("station_id").order("timestamp").execute()
    data = response.data
    df = pd.DataFrame(data)
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    return df

# Load data
df = fetch_water_levels()

data = df['water_level'].values
dates = df['timestamp'].values

# Train/test split index
split = int(len(data) * 0.7)
window = 5  # window size for residual feature creation

# 1. Fit ARIMA model on full data (or train subset ideally)
arima_order = (2,0,1)  # your optimized parameters
train_data = data[:split]
model = ARIMA(train_data, order=arima_order)
model_fit = model.fit()
arima_fitted = model_fit.fittedvalues
arima_forecast = model_fit.forecast(steps=len(data) - split)

# 2. Calculate residuals on train subset (for RF training)
residuals = train_data - arima_fitted

# 3. Prepare features for RF on residuals using sliding window
def create_features(series, window=5):
    X, y = [], []
    for i in range(len(series) - window):
        X.append(series[i:i+window])
        y.append(series[i + window])
    return np.array(X), np.array(y)

X, y = create_features(residuals, window=window)

# Train-test split for RF
rf_split = int(len(X) * 0.7)
X_train, y_train = X[:rf_split], y[:rf_split]
X_test, y_test = X[rf_split:], y[rf_split:]

# 4. Train the Random Forest on training residuals
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict residuals for test portion
rf_pred = rf.predict(X_test)

# 5. Prepare final predictions by adding ARIMA forecast + predicted residuals
# Use arima fitted values for train predictions + forecast + RF corrected residuals for test set
final_pred_train = arima_fitted[:rf_split + window]
final_pred_test = arima_forecast[:len(rf_pred)] + rf_pred

# Construct full final predictions array for plotting
final_pred_full = np.concatenate([final_pred_train, final_pred_test])

# Prepare arrays for actual water levels with corresponding indexing
actual_train = train_data[:rf_split + window]
actual_test = data[split:split + len(rf_pred)]

# Dates for plotting segments
dates_train = dates[:rf_split + window]
dates_test = dates[split:split + len(rf_pred)]

# Plot with blue = train, black = actual test, red = prediction
plt.figure(figsize=(12, 6))
plt.plot(dates_train, actual_train, label="Train", color="blue")
plt.plot(dates_test, actual_test, label="Actual", color="black")
plt.plot(dates_test, final_pred_test, label="Prediction", color="red", linestyle="--")

plt.title(f"Hybrid ARIMA({arima_order[0]},{arima_order[1]},{arima_order[2]}) + RF Forecast vs Actual")
plt.xlabel("Time")
plt.ylabel("Water Level")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Compute and print accuracy metrics on test set
mse = mean_squared_error(actual_test, final_pred_test)
rmse = np.sqrt(mse)
r2 = r2_score(actual_test, final_pred_test)
print(f"Hybrid model RMSE: {rmse}")
print(f"Hybrid model R^2: {r2}")

```
